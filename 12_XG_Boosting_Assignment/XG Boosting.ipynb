```python
# Importing necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import GridSearchCV
%matplotlib inline
```

```python
# Loading and Reading the dataset
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header = None)
test_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.test' , skiprows = 1, header = None)
```

```python
col_labels  =  ['age',  'workclass',  'fnlwgt',  'education',  'education_num', 'marital_status',  'occupation','relationship',  'race',  'sex',  'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'wage_class']
```

```python
# Setting col_labels for our train_set and test_set
train_set.columns=col_labels
test_set.columns=col_labels
```

```python
# Loading the dataset
train_set.head()
```

```python
#Loading the test_set
test_set.head()
```

```python
# Combining both the dataset
df = pd.concat([train_set,test_set])
df.head()
```

```python
# Checking for information in the dataset
df.info()
```

```python
# Checking for statistical information in the dataset
df.describe()
```

```python
# Checking for null values in the dataset
df.isnull().sum()
```

# Finding the unique values in each columns

```python
for c in df.columns:
    print(df[c].value_counts())
```

```python
# Replacing '?' with NAN values for Workclass, Occupation and native_country
```

```python
df['workclass'] = df['workclass'].replace('?',np.nan, inplace=True)
df['occupation'] = df['occupation'].replace('?',np.nan, inplace=True)
df['native_country'] = df['native_country'].replace('?',np.nan, inplace=True)
```

```python
df.isnull().sum()
```

```python
df['workclass'].fillna('0', inplace=True)
df['occupation'].fillna('0',inplace=True)
df['native_country'].fillna('0',inplace=True)
```

```python
df.isnull().sum()
```

```python
#  There is no null values in the dataset
```

```python
df.head()
```

```python
df['wage_class'].unique()
```

```python
df['wage_class'].replace('<=50K.', '<=50K', inplace=True)
df['wage_class'].replace('>50K.', '>50K', inplace=True)
```

```python
df['wage_class'].unique()
```

```python
X=df.drop(['wage_class'], axis=1)
Y=df['wage_class']
```

```python
X.columns
```

```python
X = pd.get_dummies(X)
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X_scaled,Y,test_size=0.3,random_state=355)
```

```python
X_train.shape
```

```python
Y_train.shape
```

# XG Boosting Model:

```python
from xgboost import XGBClassifier
XGB = XGBClassifier()
XGB.fit(X_train,Y_train)
```

# Hyper Parameter selection

```python
grid_param = [{ 'learning_rate':[0.1,0.001],'max_depth': [2,6,10],'n_estimators':[10,25,50]}]
```

```python
grid_search = GridSearchCV(XGB,grid_param,scoring='accuracy',cv=5,n_jobs=3,verbose=3)
grid_search.fit(X_train, Y_train)
```

```python
grid_search.best_params_
```

```python
XGB1 = XGBClassifier(learning_rate = 0.01, max_depth = 10, n_estimators = 200)
XGB1.fit(X_train,Y_train)
```

```python
Y_prediction = XGB.predict(X_test)
```
