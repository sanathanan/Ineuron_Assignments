```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random
%matplotlib inline
```

```python
# Loading the dataset and reading the first 10 records in the dataset
df = sns.load_dataset('iris')
df.head()
```

```python
X = df.drop('species', axis=1).values
Y = df['species'].values
print(X)
print(Y)
```

```python
# We are scaling the data. we are bringing the data close to mean
X = StandardScaler().fit_transform(X)
X
```

```python
# we are converting 4 dimesnions into 3 dimensions
pca = PCA(n_components=3)
pca1 = pca.fit_transform(X)
pca1 # we converted 4D into 3D
```

```python
principalDf = pd.DataFrame(data = pca1, columns = ['principal component 1', 'principal component 2', 'principal component 3'])
```

```python
# Concatenating the (X and Y) in Dataframe along axis =1. It is our final dataset for plotting
finalDf = pd.concat([principalDf, df[['species']]], axis = 1)
finalDf
```

```python
# Plotting our final dataset
```

```python
fig = plt.figure(figsize=(8, 8))

#ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
ax = fig.add_subplot(111, projection='3d')

ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_zlabel('Principal Component 3', fontsize = 15)

ax.set_title('3 component PCA', fontsize = 20)

species = ['setosa', 'versicolor', 'virginica']
colors = ['r', 'g', 'b']

for speciess, color in zip(species,colors):
    indicesToKeep = finalDf['species'] == speciess
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1'],
               finalDf.loc[indicesToKeep, 'principal component 2'],
               finalDf.loc[indicesToKeep, 'principal component 3'], c = color, s = 50)
ax.legend(species)
ax.grid()
plt.show()
```
