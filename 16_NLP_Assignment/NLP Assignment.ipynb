```python
from bs4 import BeautifulSoup 
import urllib.request 
import nltk 
from nltk.corpus import stopwords 
```

```python
response = urllib.request.urlopen('http://php.net/')
html = response.read()
soup = BeautifulSoup(html,"html5lib").get_text()
```

```python
# Separating the data based on paragraph
token_data = nltk.word_tokenize(soup)
token_data
```

```python
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
stop_words
```

```python
# We also have punctuations which we can ignore from our set of words just like stopwords.

import string

punct =string.punctuation
punct
```

```python
# Let's word tokenize the given sample after we remove the stopwords and punctuation. 

import nltk
import string
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
punct =string.punctuation

raw_data= token_data
clean_data =[]
for word in raw_data:
    if word not in punct:
        if word not in stop_words:
            clean_data.append(word)
            
clean_data
```

```python
# Finding the frequency of words

freq_words = nltk.FreqDist(clean_data)

for key,val in freq_words.items(): 
    print (str(key) + ':' + str(val))
```
