```python
# Importing necessary Libraries
import numpy as np
import scipy
from scipy import misc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loading and visualizing the gray scale image
racoon = scipy.misc.face(gray=True)
plt.figure(figsize=(10,15))
plt.imshow(racoon, cmap=plt.cm.gray)
plt.show()
```

```python
# Looking for the size of the image
rows = racoon.shape[0]
cols = racoon.shape[1]
image = racoon.reshape(rows*cols,1)
```

```python
from sklearn.cluster import KMeans
k_mean = KMeans(n_clusters=5)
k_mean.fit(image)
```

```python
clusters = np.asarray(k_mean.cluster_centers_) 
labels = np.asarray(k_mean.labels_)  
labels = labels.reshape(rows,cols) 
```

```python
# Saving the Image
plt.imsave('compressed_racoon.png',labels)
```

```python
# Visualizing the image
clustered_Image = plt.imread('compressed_racoon.png')
plt.figure(figsize=(10,15))
plt.imshow(clustered_Image)
plt.show()
```
